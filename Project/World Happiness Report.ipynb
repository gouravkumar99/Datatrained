```python
#import python library used for analysis data
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create variable and store all data into variable using pandas 
data = pd.read_csv(r'C:\Users\goura\Downloads\DSData-master\DSData-master\happiness_score_dataset.csv')
```

```python
#print 1st 5 rows
data.head()
```

```python
#create list all need data into list 
data_columns = ['Country','Region','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Generosity','Trust (Government Corruption)']
```

```python
happiness_df = data[data_columns].copy()
```

```python
#print first 5 rows 
happiness_df.head()
```

```python
#check all null values column wise in data frame 
happiness_df.isnull().sum()
```

```python
#happiness score and GDP
plt.rcParams['figure.figsize'] = (15,7)
plt.title('Happiness score and GDP',fontsize=20)
sns.scatterplot(x=happiness_df['Happiness Score'],y=happiness_df['Economy (GDP per Capita)'], hue=happiness_df['Region'],s=200)
plt.xlabel('Happiness Score',fontsize=20)
plt.ylabel('Economy (GDP per Capita)',fontsize=20)
plt.rcParams['figure.facecolor']='#ADD8E6'
```

```python
#loest happines Sub-Saharan Africa
#Higest happines Western Europe
```
